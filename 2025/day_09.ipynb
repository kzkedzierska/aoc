## Part 1

While elves are redecorating and for some reason they need the largest rectangle created between the two red tiles across a diagonal. The input is the position of the red files. 

```python
import helper as h
from pathlib import Path
logger = h.setup_logger("AoC_2025: Day 9")
```

```python
def read_input(src: str | Path):
    if Path(src).is_file():
        with open(src, "r") as f:
            data = f.read().strip().splitlines()
    else:
        data = src.strip().splitlines()

    return [tuple([int(el) for el in line.split(",")]) for line in data]
```

```python

def rect_area(rt1, rt2):
    return (abs(rt1[0] - rt2[0])+1) * (abs(rt1[1] - rt2[1])+1)

def largest_square(red_tiles: list[tuple[int, int]]):
    biggest_area = -1

    for i in range(len(red_tiles)):
        for j in range(i + 1, len(red_tiles)):
            area = rect_area(red_tiles[i], red_tiles[j])
            if area > biggest_area:
                biggest_area = area


    return biggest_area

def part1(src: str | Path):
    red_tiles = read_input(src)
    return largest_square(red_tiles)

h.extended_assert("toy", part1("inputs/day_09_toy.txt"), 50)
```
