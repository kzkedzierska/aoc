## Part 1

We need to open the safe, for which we have instructions how to turn the dial. The code to the safe is the number of times the dial landed at 0 at the end of a movement. The dial moves from 0 to 99. If we turn the dial right, we add to it. If we cross 99 while adding up, we start from 0. (i.e., starting at 99 we add 3, we end up with 2). If we turn left, we subtract. If we cross 0 while subtracting, we subtract from 100, i.e., starting form 1, we subtract 5, we end up at 96.

```python
import helper as h
logger = h.setup_logger("AoC_2025: Day 1")

logger.info("First day of the AoC 2025!")
```

```python
from pathlib import Path
def read_input(input: str | Path):
    if Path(input).is_file():
        with open(input, 'r') as f:
            data = f.read().strip().splitlines()
    else:
        data = input.strip().splitlines()
    
    return [float(line.replace("R", "+").replace("L", "-")) for line in data]
```

```python
def part1(input, start_with=50):
    data = read_input(input)
    n_zeros = 0
    current = start_with
    for d in data:
        current = (current + d) % 100
        n_zeros += current == 0
    return n_zeros

h.verify_answer(part1, "./inputs/day_01_toy.txt", 3)
```

```python
h.verify_answer(part1, "inputs/day_01.txt", 1177)
```

## Part 2

The modification is that now we need to count how many times the dial crosses the 0 or lands on the 0. If it landed on 0 in the previous round, going to -5 would not be considered crossing the 0. 


```python
def n_cross_zero(start, move):
    n_rots = abs(move) // 100 # full rotations
    move = move - n_rots * 100 if move > 0 else move + n_rots * 100
    n_rots += abs((start + move) // 100) if start > 0 else 0
    n_rots += (start + move) == 0 if start > 0 else 0
    return n_rots
```

```python
h.extended_assert(n_cross_zero(0, -5), 0)
h.extended_assert(n_cross_zero(50, 49), 0)
h.extended_assert(n_cross_zero(49, 1000), 10)
h.extended_assert(n_cross_zero(99, 100), 1)
h.extended_assert(n_cross_zero(0, 100), 1)
h.extended_assert(n_cross_zero(0, 300), 3)
h.extended_assert(n_cross_zero(49, 51), 1)
h.extended_assert(n_cross_zero(0, 0), 0)
```

```python
h.extended_assert(n_cross_zero(50, -68), 1)
h.extended_assert(n_cross_zero(82, -30), 0)
h.extended_assert(n_cross_zero(52, 48), 1)
h.extended_assert(n_cross_zero(0, -5), 0)
h.extended_assert(n_cross_zero(95, 60), 1)
h.extended_assert(n_cross_zero(55, -55), 1)
h.extended_assert(n_cross_zero(0, -1), 0)
h.extended_assert(n_cross_zero(99, -99), 1)
h.extended_assert(n_cross_zero(0, 14), 0)
h.extended_assert(n_cross_zero(14, -82), 1)
```

```python
def part2(input, start_with=50):
    data = read_input(input)
    n_zeros = 0
    current = start_with
    for d in data:
        n_zeros += n_cross_zero(current, d)
        current = (current + d) % 100
    return n_zeros

h.verify_answer(part2, "./inputs/day_01_toy.txt", 6)
```
