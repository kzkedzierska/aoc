## Part 1

We're about to solve a very weird math problem. The elements of the equation are in a column, and each equation is separated by a blank line. The operation to be performed is indicated by the *last* element of each equation.

````python
import helper as h
from pathlib import Path
import re
logger = h.setup_logger("AoC_2025: Day 6", level="INFO")
````

````python
def part1(src: str | Path):
    if Path(src).is_file():
        with open(src, "r") as f:
            data = f.read().strip().splitlines()
    else:
        data = src.strip().splitlines()
    
    total = 0
    s_ = []
    m_ = []
    for line in data:
        line = line.strip()
        line = re.sub(r"\s+", " ", line)
        logger.debug(f"{line[:5]}")

        if line[0] == "+" or line[0] == "*":
            signs = line.split(" ")
            total = [s_[i] if signs[i] == "+" else m_[i] for i in range(len(signs))]
            logger.debug(f"{total=}")
            return sum(total)

        nums = [int(n) for n in line.split(" ")]
        logger.debug(f"{nums[:5]}")
        if len(s_) == 0:
            s_.extend([0] * len(nums))
            m_.extend([1] * len(nums))
        # elif len(s_) < len(nums): # in case the number of columns is not constant?
        # i checked with `cat day_06.txt | tr -s " " | awk '{print NF}' | sort | uniq`
        #     # add 0s for sums and 1s for multiplications
        #     s_.extend([0] * (len(nums) - len(s_)))
        #     m_.extend([1] * (len(nums) - len(m_)))

        s_ = [s_[i] + nums[i] for i in range(len(nums))]
        m_ = [m_[i] * nums[i] for i in range(len(nums))]

        
h.extended_assert("toy", part1("inputs/day_06_toy.txt"), 4277556)
````

## Part 2

Now, each number is written from top to bottom, right-to-left, in a column. What a messy math. 


```
789
56
1234
```

I think the above would be written into: 

```
4, 93, 862, 751
```

````python
def read_input_p2(src: str | Path):
    if Path(src).is_file():
        with open(src, "r") as f:
            data = f.read().strip().splitlines()
    else:
        data = src.strip().splitlines()

    num_strings = ["" for _ in range(len(data[0]))]
    for line in data[:-1]:
        for i, char in enumerate(reversed(line)):
            
            num_strings[i] = num_strings[i] + char
    
    eqs = re.sub(r"\s+", " ", data[-1][::-1].strip()).split(" ")

    return num_strings, eqs


dummy_test_input = """
789 
56  
1234
+
"""
h.extended_assert("dummy_test", read_input_p2(dummy_test_input), (['  4', '9 3', '862', '751'], ['+']))
````

````python
def process_equations(num_strings, eqs):
    total = 0
    i = 0
    teq = 0 if eqs[i] == "+" else 1
    for num_s in num_strings:
        num = re.sub(r"\s+", " ", num_s)
        try:
            num = int(num)
            if eqs[i] == "+":
                teq += num

            else:
                teq *= num

        except ValueError:
            logger.debug(f"{teq=}")
            total += teq
            i += 1
            teq = 0 if eqs[i] == "+" else 1
    
    return total + teq

def part2(src: str | Path):
    return process_equations(*read_input_p2(src))

h.extended_assert("toy", part2("inputs/day_06_toy.txt"), 3263827)
````
